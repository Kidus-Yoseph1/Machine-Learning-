#### Remove columns that have more than 25% null 

```python
def remove_cols_with_high_nulls(df, threshold = 25):
    # Percentage of null values in each column 
    null_percentages = df.isnull().sum() / len(df) * 100
    
    # identify cols to drop based on the threshold
    columns_to_drop = null_percentages[null_percentages >= threshold].index.tolist()
    
    # drop identifeid cols
    df_cleaned = df.drop(columns=columns_to_drop)

    if columns_to_drop:
        print(f"Columns removed due to high null percentage ({threshold}%): {columns_to_drop}")
        print(f"New DataFrame shape: {df_cleaned.shape}")
    else:
        print(f"No columns removed with null percentage >= {threshold}%.")

    return df_cleaned, columns_to_drop
    
```

```python
df_cleaned , columns_to_drop = remove_cols_with_high_nulls(df.copy(),threshold = 25 )
df_cleaned
```

### Remove Rows with null values 

```python
df_cleaned = df_cleaned.dropna()
df_cleaned.isnull().any()
```
